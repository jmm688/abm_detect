```python
import rasterio as rio
import numpy as np
import pandas as pd
import seaborn as sns
import os 

import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
```

```python
### Opening raster and extracting np.array

def get_array(path):

    rast = rio.open(path)
    image = rast.read()
    rast.close()
    
    return image

```

```python
mine = '/home/jmm688/Pictures/pictures/discussion/SVC/gtifs/mines/prob_1/mine__3.tif'
mine_color = '/home/jmm688/Pictures/discussion_mine_regs.tif'
labels = '/home/jmm688/Pictures/pictures/discussion/linreg/gtifs/mines/labels/mine__3.tif'

im = get_array(mine)
im_2 = get_array(mine_color)
label_im = get_array(labels)
```

```python
sns.set(font_scale=4)
scalebar = ScaleBar(1,
                        location = 'upper right',
                        pad = 0.9,border_pad = 0.1,
                        length_fraction = .5,
                        box_alpha = 1,
                        font_properties = {'size': 'xx-large','weight' : 'roman'}) # 1 pixel = 0.2 meter

fig, axes = plt.subplots(figsize=(25,25))
sns.heatmap(im[0,14:34,15:35],cbar= True,square = True, annot=False, cmap="plasma_r",annot_kws={"size": 25},cbar_kws={"label": '\nPLSR-C    Decision Function','shrink':.8},xticklabels=False, yticklabels = False)
p = sns.heatmap(im_2[11,15:35,15:35],cbar= False,square = True,annot=True, cmap="plasma_r",annot_kws={"size": 20},xticklabels=False, yticklabels = False, alpha=.1,fmt='g')

axes.add_artist(scalebar)



plt.show()
```
