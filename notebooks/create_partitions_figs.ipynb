# Generating meaningful visuals from output data

```python
'''
The purpose of this notebook is to generate graphs from partition output in order to insert in results thesis section
this way the spreadsheet can go in the appendix instead of within the results section
'''
```

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Linear Regression scores

```python
# loading in a data set to make a test run on images
linreg_path ='/home/jmm688/github/abm_detect/modeling/test_workspace_thesis/Linear_regression/linreg.csv'

# get df
lreg_df = pd.read_csv(linreg_path,index_col=0)
# get averages for scores
average_reg = lreg_df['Regression Score'].mean()
average_class = lreg_df['Classification score'].mean()

# boxplot stuff
sns.set_theme(style="whitegrid")
plt.figure(figsize = (10,4))
ax_0 = sns.boxplot(data = lreg_df[['Regression Score','Classification score']],orient='h',width=.7)
# boxplot text
plt.text(75.0, 1.1, "Classification Average\n"+format(average_class,'.1f')+'%', horizontalalignment='center', size=11, color='black', weight='semibold')
plt.text(61.5, 0.1, "Regression Average\n"+format(average_reg,'.1f')+'%', horizontalalignment='center', size=11, color='black', weight='semibold')



plt.title('Linear Regression model performance\n on all partitions at TM',fontsize=15)
```
